```python
import pandas as pd
```

```python
dataset = pd.read_excel("myexcelproject.xlsx")
```

```python
dataset
```

```python
import numpy as np

# Replace the 'Height' column with random values
np.random.seed(42)
dataset['Height'] = np.random.randint(150, 181, size=len(dataset))

# Verify the changes
display(dataset.head())
```

```python
team_distribution = dataset['Team'].value_counts()
team_percentage = (team_distribution / len(dataset)) * 100

# Combine results into a DataFrame
team_summary = pd.DataFrame({
    'Count': team_distribution,
    'Percentage': team_percentage
})
display(team_summary)
```

```python
import matplotlib.pyplot as plt

# Pie chart
team_distribution.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Employee Distribution by Team')
plt.ylabel('')
plt.show()
```

```python
position_distribution = dataset['Position'].value_counts()
display(position_distribution)
```

```python
position_distribution.plot(kind='bar', color='skyblue')
plt.title('Employee Distribution by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

```python
age_bins = [18, 25, 35, 45, 55, 65]
dataset['Age Group'] = pd.cut(dataset['Age'], bins=age_bins)
age_group_distribution = dataset['Age Group'].value_counts()
display(age_group_distribution)
```

```python
age_group_distribution.plot(kind='bar', color='coral')
plt.title('Employee Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
team_salary = dataset.groupby('Team')['Salary'].sum()
position_salary = dataset.groupby('Position')['Salary'].sum()

# Find the highest salary expenditures
highest_team = team_salary.idxmax()
highest_position = position_salary.idxmax()
print(f"Highest Salary by Team: {highest_team} (${team_salary[highest_team]:,.2f})")
print(f"Highest Salary by Position: {highest_position} (${position_salary[highest_position]:,.2f})")
```

```python
team_salary.sort_values().plot(kind='barh', color='lightgreen')
plt.title('Total Salary by Team')
plt.xlabel('Total Salary')
plt.show()
```

```python
correlation = dataset[['Age', 'Salary']].corr()
print("Correlation between Age and Salary:")
print(correlation)

# Scatter plot
plt.scatter(dataset['Age'], dataset['Salary'], color='purple', alpha=0.6)
plt.title('Age vs Salary Correlation')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
